```python
import spacy

# load English Large Model as nlp objects
nlp = spacy.load("en_core_web_lg")
```

```python
# testing the word, that has vector or oov(out of vector) 
doc = nlp("cat mango car tanvir")

for token in doc:
    print(f"{token.text} | vector ==> {token.has_vector} | out of vector ==> {token.is_oov}")
```

# Get Similarity score between two words using Spacy English Large Model

```python
based_word = nlp("bread")

words_for_test = nlp("bread sandwich burger car tiger human wheat")
for token in words_for_test:
    print(f"{token.text} <==> {based_word.text} :", token.similarity(based_word))
```

```python
# using fuction
def similarity_score(based_word, words_to_compare):
    based_token = nlp(based_word)
    doc = nlp(words_to_compare)
    
    for token in doc:
        print(f"{token.text} <==> {based_token.text} :", token.similarity(based_token))
```

```python
similarity_score("iphone", "apple samsung iphone cat banana oppo")
```

# Get Similarity score between two words using Scikit-Learn Cosine Similarity

```python
# converting word to vector
king = nlp.vocab["king"].vector
king
```

```python
# converting word to vector
king = nlp.vocab["king"].vector
queen = nlp.vocab["queen"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector

result = king - man + woman
result
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity([result], [queen])
```
