```python
# Creating stop words and punctuation processing function
def processing(text):
    
    doc = nlp(text)
    
    without_stopwords = [token.text for token in doc if not token.is_stop and not token.is_punct]
    
    # Converting list to string.We need string,so that process
    return " ".join(without_stopwords)
```

```python
# creating new processed column
new_df["processed_short_description"] = new_df["short_description"].apply(processing)
new_df
```
